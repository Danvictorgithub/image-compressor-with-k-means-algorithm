```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
original_image = plt.imread('./image.jpeg')
original_image.shape

# Feature scale image
original_image = original_image/ 255

# Combines pixels pos(x, y) and RGB values into a single array
X_img = np.reshape(original_image,(original_image.shape[0]*original_image.shape[1],3))
X_img[1]
```

```python
# Attempts to convert 24-bits RGB values to K-bits RGB values
K = 8
model = KMeans(n_clusters=K)
model.fit(X_img)

# Assigns each pixel to a cluster
idx = model.cluster_centers_[model.predict(X_img)]

# Reshapes the image
compressed_image = np.reshape(
    idx, 
    (original_image.shape[0], original_image.shape[1], 3)
    )
```

```python
# Display original image
fig, ax = plt.subplots(1,2, figsize=(8,8))
plt.axis('off')

ax[0].imshow(original_image)
ax[0].set_title('Original')
ax[0].set_axis_off()


# Display compressed image
ax[1].imshow(compressed_image)
ax[1].set_title('Compressed with %d colours'%K)
ax[1].set_axis_off()
```
